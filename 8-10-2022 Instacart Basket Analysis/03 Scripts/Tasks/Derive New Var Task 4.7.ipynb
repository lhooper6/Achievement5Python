```python
#  2 Looka t the first 1 mil rows 
df = dforders_products_merged[:1000000]
```

```python
# Create Conditions
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# 3 Define and apply price label fucntion to subset 
df['price_range'] = df.apply(price_label, axis=1)
```

# Task 4.7

saturday and sunday are the busiest days (0 & 1).  Tuesday and Wednesday are the least busiest two days (3 & 4).


```python
# 2 create a new loop
result1 = []

for value in dforders_products_merged["order_dow"]:
  if value == 0 or value == 1:
    result1.append("Busiest days")
  elif value == 3 or value == 4:
    result1.append("Least busy days")
  else:
    result1.append("Regularly busy")
```

```python
# add to dataframe with correct label 
dforders_products_merged['busiest_days'] = result1
```

```python
# 3 check new column for accuracy
dforders_products_merged['busiest_days'].value_counts(dropna = False)
```

These new values make sense.  I noted that majority of orders happen on the regularly busy days with the busiest days closely behind.  This would make sense because the regular busy days consist of 3 days: Monday, Thursday, and Friday.

```python
# 4 Find the column I want to use
dforders_products_merged.columns
```

```python
# 4 find out the busiest hours
dforders_products_merged['order_hour_of_day'].value_counts(dropna = False)
```

Based on these hours, I would assign hours 0-6 as the least orders, 9-16 as most orders, and the rest as average orders.

```python
# 4 create a new loop
result2 = []

for value in dforders_products_merged['order_hour_of_day']:
  if value>=9 and value<=16:
    result2.append("Most Orders")
  elif value<=6:
    result2.append("Fewest Orders")
  else:
    result2.append("Average Orders")
```

```python
# 4 add to dataframe with correct label 
dforders_products_merged['busiest_period_of_day'] = result2
```

```python
# 5 check new column for accuracy
dforders_products_merged['busiest_period_of_day'].value_counts(dropna = False)
```
