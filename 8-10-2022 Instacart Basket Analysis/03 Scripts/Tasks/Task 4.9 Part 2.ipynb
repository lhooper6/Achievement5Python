# 3 Create and Analyze Histogram

```python
# Create Histogram
All_data['order_hour_of_day'].plot.hist(bins = 25)
plt.title('Distribution of Order Hour of Day')
```

```python
# change bins 
hist = All_data['order_hour_of_day'].plot.hist(bins = 70)
plt.title('Distribution of Order Hour of Day')
```

The histogram shows that the majority of the orders happen during the day hours of 10-15 and drop of more in the evening and early morning. 

# 4 Create and Analyze Bar Chart

```python
# create bar chart 
Bar = All_data['loyalty_flag'].value_counts().plot.bar()
plt.title('Loyalty Flag')
plt.xlabel('count')
```

# 5 Create and Analyze Line Chart

```python
# create seed for sample 
np.random.seed(5)
dev = np.random.rand(len(All_data)) <= 0.7
```

```python
# Split data frame into 2 samples
# 70%
big = All_data[dev]
```

```python
# 30%
small = All_data[~dev]
```

```python
# check they equal the whole data frame 
len(All_data)
```

```python
len(big) + len(small)
```

```python
# Reduce samples to applicable columns 
df_line = small[['order_hour_of_day', 'prices']]
```

```python
# Make line graph
line = sns.lineplot(data = df_line, x = 'order_hour_of_day', y = 'prices')
plt.title('Order Hour of Day and Prices')
```

There are some differences in expenditure depending on the hour of the day. It seems that higher price items are purchased more in the ealrier hours 0-5.

# 6 Line Chart - age and dependents

```python
# Create a new df with applicable columns so you can use the same sample
df_line2 = small[['Age', 'number_dependants']]
```

```python
# Make line graph
line2 = sns.lineplot(data = df_line2, x = 'Age', y = 'number_dependants')
plt.title('Age and Number of Dependants')
```

There is not much of a connection between age and the number of dependents.  This is evidenced by the rise and fall of data points at every differnt age.  

# 7 Create a Scatter for Income and Age

```python
# Create a new df for sactter
df_scatter = small[['Age', 'income']]
```

```python
scatter = sns.scatterplot(data = df_scatter, x = 'Age', y = 'income')
plt.title('Age and Income')
```

The scatterplot shows that as the customer ages, their income seems to be higher.  The bulk of the points increase shortly after age 40 menaing a 40 year old would have more spending power than a 20 year old.
