```python
# Import Libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import Data
path = r'C:\Users\419182\8-10-2022 Instacart Basket Analysis'
```

```python
# 1 Import Data
ords_prods_merge_3 = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_3.pkl'))
```

```python
# 2 Find the aggregated mean of order number grouped by department id
ords_prods_merge_3.groupby('department_id').agg({'order_number': ['mean']})
```

3 Analyze Result: The means for the subset of data and the means of the popultation do and shoud differ slightly.  In the subset, only 100000 rows with orders are available.  When using the population data set, the mean represents all orders in a specific department.  

```python
# 4 Create Loyalty Flag: create new column
ords_prods_merge_3['max_order'] = ords_prods_merge_3.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#  check
ords_prods_merge_3.head(10)
```

```python
# Create flag based on max order 
ords_prods_merge_3.loc[ords_prods_merge_3['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge_3.loc[(ords_prods_merge_3['max_order'] <= 40) & (ords_prods_merge_3['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge_3.loc[ords_prods_merge_3['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge_3['loyalty_flag'].value_counts(dropna = False)
```

```python
# 5 Check statistics of customers based on layalty flag
ords_prods_merge_3.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

```python
# 6 Create spending flag: create a mean order price column
ords_prods_merge_3['mean_order'] = ords_prods_merge_3.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge_3.shape
```

```python
ords_prods_merge_3.head(5)
```

```python
# 6 create flag based on price
ords_prods_merge_3.loc[ords_prods_merge_3['mean_order']<10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge_3.loc[ords_prods_merge_3['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
# 6 check 
ords_prods_merge_3['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge_3[['mean_order', 'spending_flag']].head(100)
```

```python
ords_prods_merge_3.columns
```

```python
# 7 Create a flag based on days since prior order: create column showing median
ords_prods_merge_3['median_days_since_prior_order'] = ords_prods_merge_3.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check
ords_prods_merge_3.shape
```

```python
# 7 create flag
ords_prods_merge_3.loc[ords_prods_merge_3['median_days_since_prior_order']>20, 'Frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge_3.loc[(ords_prods_merge_3['median_days_since_prior_order']>10) & (ords_prods_merge_3['median_days_since_prior_order']<=20), 'Frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge_3.loc[ords_prods_merge_3['median_days_since_prior_order']<=10, 'Frequency_flag'] = 'Frequent customer'
```
